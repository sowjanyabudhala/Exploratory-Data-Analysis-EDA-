```python
numerical_columns=['PassengerId','Survived','Pclass','Age','SibSp','Parch','Fare']
for columns in numerical_columns:
  sns.boxplot(x=df[columns],color='black')
  plt.title(f'Boxplot of {columns}')
  plt.show()
```

```python
numerical_columns=['PassengerId','Survived','Pclass','Age','SibSp','Parch','Fare']
for columns in numerical_columns:
  sns.kdeplot(x=df[columns],color='red')
  plt.title(f'kdeplot of {columns}')
  plt.show()
```

**conclusion**:this plot shows the skewness of the data survived ,Age and pclass are normally distriuted .sibsp,parch,fare are right skewed.

```python
numerical_columns=['PassengerId','Survived','Pclass','Age','SibSp','Parch','Fare']
for columns in numerical_columns:
  sns.distplot(x=df[columns],color='red')
  plt.title(f'distplot of {columns}')
  plt.show()
```

pairplot/correlation matrix for feature relationships.

```python
numerical_columns=['PassengerId','Survived','Pclass','Age','SibSp','Parch','Fare']
sns.pairplot(df[numerical_columns])
plt.show()
```
